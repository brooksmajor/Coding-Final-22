```python
#@title Image 1
#Setup such as speed of the drawing, colors, size of drawing
clearscreen()
#Have to import random since it the drawing is randomly generated
import random
setup(500,500)
hideturtle()
speed(13)
bgcolor("black")
pencolor("white")

#Create a list to capture (x,y) data
telemetry_list = []

#Create a tiling function
#Starts at l starts at 0 beginging of recursion
def tiling(x,y,s,l):
  if l == 0:
#draws two dots vertically
    #If the random number is below 0.5 it will do the vertical part if it isn't than it will exicute the horizonatal part of the function
    if random.random() < 0.5:
      penup()
      goto(x,y-s)
      dot(2)
      coordinates = (x, y-s)
      #capturing the (x,y) data to put into the telemetry_list
      telemetry_list.append(coordinates)
      goto(x,y+s)
      dot(5)
      coordinates = (x, y+s)
      telemetry_list.append(coordinates)
#draws two dots horizontally
    else:
      penup()
      goto(x-s,y)
      coordinates = (x-s, y)
      telemetry_list.append(coordinates)
      dot(10)
      goto(x+s,y)
      coordinates = (x+s, y)
      telemetry_list.append(coordinates)
      dot(25)
# S is split screen and go to next level of recursion
# Displayed below is the tiling 4 times 
  else:
    s /= 2
    l -= 1
    tiling(x-s,y+s,s,l)
    tiling(x+s,y+s,s,l)
    tiling(x-s,y-s,s,l)
    tiling(x+s,y-s,s,l)
  return telemetry_list 
#Capture the image
screen_obj = Screen()

# Ex: data = tiling(x,y,s,l)
# for l the lower the number the less amount of dots
# for s the scale is increased or decreased
# tiling(0,0,100,2) and captuyring the data that is tiled

data = tiling(0,0,100,2)

#saving the dictionary data as json file
import json
with open("Image_1.json", 'w') as json_obj:
  json.dump(data, json_obj, indent=2)
print('file written')

#saving the image as an SVG file
screen_obj.saveSVG('Image_1.svg', False)
```

```python
#@title Image 3
#Setup such as speed of the drawing, colors, size of drawing
clearscreen()
import random
setup(500,500)
hideturtle()
speed(13)
bgcolor("black")
pencolor("white")

telemetry_list = []

#Straight function
def tiling(x, y, s, l, mode="straight"):
  if l == 0:
    if mode == 'straight':
      if random.random() < 0.5:
        penup()
        goto(x,y-s)
        dot(2)
        coordinates = (x, y-s)
        telemetry_list.append(coordinates)
        goto(x,y+s)
        dot(5)
        coordinates = (x, y+s)
        telemetry_list.append(coordinates)

      else:
        penup()
        goto(x-s,y)
        dot(10)
        coordinates = (x-s,y)
        telemetry_list.append(coordinates)
        goto(x+s,y)
        dot(25)
        coordinates = (x+s,y)
        telemetry_list.append(coordinates)
#Created another function within the function to make a diagonal pattern
#The only diffrence is that the path of drawing the dots is at an angle. 
#This is shown within the x, y values as they are both being subtracted or added or both.
    elif mode == 'diagonal':
        if random.random() < 0.5:
          penup()
          goto(x-s,y+s)
          coordinates = (x-s, y+s)
          telemetry_list.append(coordinates)
          dot(5)
          goto(x+s,y-s)
          coordinates = (x+s, y-s)
          telemetry_list.append(coordinates)
          dot(5)
        else:
          penup()
          goto(x-s,y-s)
          coordinates = (x-s, y-s)
          telemetry_list.append(coordinates)
          dot(15)
          goto(x+s,y+s)
          coordinates = (x+s, y+s)
          telemetry_list.append(coordinates)
          dot(5)
 #I decided to experiment and wanted to divide s by 3 instead of 2 
 # This seperates the sections that are being repeated/oriented
  else:
    s /= 3
    l -= 1
    tiling(x-s,y+s,s,l,mode)
    tiling(x+s,y+s,s,l,mode)
    tiling(x-s,y-s,s,l,mode)
    tiling(x+s,y-s,s,l,mode)
  return telemetry_list 
screen_obj = Screen()

# tiling((0,0,200,3, mode='diagonal') it could also be the srtaight mode to get a different output image
data = tiling(0,0,200,3, mode='diagonal')

#save dictionary data as json file
import json
with open("Image_3.json", 'w') as json_obj:
  json.dump(data, json_obj, indent=2)
print('file written')

#save the image as an SVG file
screen_obj.saveSVG('Image_3.svg', False)
```
